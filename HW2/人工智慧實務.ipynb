```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hwweiCOAIPP/test/main/SimpleWeather.csv")
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
no = Normalizer()
X = no.fit_transform(df.drop(['Temperature (C)'],axis = 1))
Y = df['Temperature (C)']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)
lr.fit(X_train,Y_train)
lr.score(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection ="3d")
x=X[1:100,0]
y=X[1:100,5]
z=Y[1:100]
ax.set_xlabel("Unnamed: 0")
ax.set_ylabel("Temperature (C)")
ax.set_zlabel("Apparent Temperature (C)")
ax.scatter(x,y,z,c=z,cmap="Reds",label="original price",marker="o")
```
